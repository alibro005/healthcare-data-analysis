- The two things that seen from info that the date of Addmision and Discharge date is str type we need to convert it into the Datetime

```python
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])
```

```python
print((df['Date of Admission']).dtypes)
print((df['Discharge Date']).dtypes)
```

```python
df.head(10)
```

```python
cols = ['Insurance Provider','Room Number']

df.drop(cols, axis=1, inplace=True)
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Length of Stay'] = (
    df['Discharge Date'] - df['Date of Admission']
).dt.days
```

```python
df.head(10)
```

```python
df["Age"].max()
```

```python
df.info()
```

```python

plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=30, color='skyblue', edgecolor='black')
plt.title('Patient Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Insight:
Most patients are concentrated in a certain age range; few very young or very old.

```python
gender_percent = (df['Gender'].value_counts(normalize=True) * 100).round(2)

plt.figure(figsize=(6,4))
plt.bar(gender_percent.index, gender_percent.values, color=['#ff9999','#66b3ff'])
plt.title('Gender Distribution (%)')
plt.ylabel('Percentage of Patients')

plt.show()
```

## Insight:

Shows balance or skew between male and female patients.

```python
top_conditions = df['Medical Condition'].value_counts().head(10)
plt.figure(figsize=(10,5))
plt.bar(top_conditions.index, top_conditions.values, color='orange', edgecolor='black')
plt.title('Top 10 Medical Conditions')
plt.xlabel('Medical Condition')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Insight:

Highlights the most frequent causes for hospitalization.

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Length of Stay'], df['Billing Amount'], alpha=0.5, color='green')
plt.title('Length of Stay vs Billing Amount')
plt.xlabel('Length of Stay (days)')
plt.ylabel('Billing Amount')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

## Insight:

Longer stays tend to lead to higher billing — strong financial driver.

```python
monthly_admissions = df.set_index('Date of Admission').resample('MS').size()

plt.figure(figsize=(10,5))
plt.plot(monthly_admissions.index, monthly_admissions.values, marker='o', color='purple')
plt.title('Monthly Admissions Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Admissions')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```


## Insights
Reveals seasonality or trends in hospital admissions over time.
